```python
from dotenv import load_dotenv
import os
import xmlrpc.client
import pandas as pd


url = os.getenv("url")
db = os.getenv("db")
username = os.getenv("username")
password = os.getenv("password")




common = xmlrpc.client.ServerProxy(f'{url}/xmlrpc/2/common')

uid = common.authenticate(
    db,
    username,
    password,
    {}
)

# MODELS
models = xmlrpc.client.ServerProxy(
    f'{url}/xmlrpc/2/object'
)

lineas = models.execute_kw(
    db,
    uid,
    password,
    'account.move.line',
    'search_read',
    [[
        ['move_id.move_type', 'in', [
            'out_invoice',
            'out_refund'
        ]],
        # ['product_id', '!=', False]
    ]],
    {
        'fields': [
            'company_id',
            'invoice_date',
            'move_id',
            'product_id',
            'partner_id',
            'account_id',
            'quantity',
            'price_unit',
            'currency_id',
            'price_subtotal',
            'balance',
            'date',
            'name',
            'analytic_distribution',
            'matching_number'
            

        ],
        'limit': 1000
    }
)
```

```python

print(lineas)
df = pd.DataFrame(lineas)
df['partner_id'] = df['partner_id'].str[1]
df['account_id'] = df['account_id'].str[1]
df['company_id'] = df['company_id'].str[1]
df['move_id'] = df['move_id'].str[1]
```
