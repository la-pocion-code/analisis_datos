```python
# Cargar el archivo Excel
ventas_df = pd.read_excel("explocion.xlsx")  # Reemplaza "ventas.xlsx" con la ruta de tu archivo
```

```python
# Esta linea de código valida los KIT que faltan tener en cuenta en el listado
set(ventas_df[ventas_df['PRODUCTO'].str.contains('KIT ')]['PRODUCTO'].drop_duplicates().reset_index(drop=True)) - set(data_kits.keys())
```

```python
# Crear una lista para almacenar las filas explotadas
filas_explodidas = []

# Iterar sobre cada fila de ventas
for _, fila in ventas_df.iterrows():
    producto = fila["Producto"]
    cantidad = fila["Cantidad"]
    
    # Verificar si el producto es un kit
    if producto in data_kits:
        # Obtener los productos del kit
        productos_kit = data_kits[producto]
        
        # Crear una fila por cada producto del kit
        for prod in productos_kit:
            fila_explodida = fila.copy()
            fila_explodida["Producto"] = prod
            fila_explodida["Cantidad"] = cantidad  # La cantidad se mantiene igual para cada producto del kit
            filas_explodidas.append(fila_explodida)
    else:
        # Si no es un kit, mantener la fila original
        filas_explodidas.append(fila)

# Crear un nuevo DataFrame con las filas explotadas
ventas_explodidas_df = pd.DataFrame(filas_explodidas)
```

```python
ventas_explodidas_df.to_excel("ventas_explodidas.xlsx", index=False)
```
