```python
import pandas as pd
import time

# 1. Configuración de archivos
archivo_adjunto = r'D:\Downloads\Comunicado cambio de razon social y nuevos datos bancarios 290126.pdf'
banner = r'D:\Downloads\image.jpeg'

# 2. Carga de correos
correos = pd.read_csv(r'D:\Desktop\correos.csv')
correos_cco = correos['Correos'].tolist()

# 3. Definir el tamaño de cada grupo (bloques de 300 para mayor seguridad)
tamaño_bloque = 300 

# Función para dividir la lista en partes
def dividir_lista(lista, n):
    for i in range(0, len(lista), n):
        yield lista[i:i + n]

# 4. Iterar y enviar por grupos
bloques = list(dividir_lista(correos_cco, tamaño_bloque))
print(f"Se han generado {len(bloques)} bloques de envío.")

for i, grupo in enumerate(bloques):
    print(f"Enviando bloque {i+1} de {len(bloques)} ({len(grupo)} destinatarios en CCO)...")
    
    exito, mensaje = ms.enviar_correo(
        destinatarios=['mburgos@lapocion.com'],
        bcc=grupo,
        asunto="Cambio de razón social y actualización de datos bancarios",
        cuerpo_html=cuerpo_html,
        adjuntos=[archivo_adjunto],
        banner=banner
    )
    
    if exito:
        print(f"✅ Bloque {i+1} enviado.")
    else:
        print(f"❌ Error en bloque {i+1}: {mensaje}")

    # Pausa de seguridad para evitar bloqueos por velocidad de envío
    if i < len(bloques) - 1:
        print("Esperando 10 segundos para el siguiente bloque...")
        time.sleep(10)

print("Proceso de envío finalizado.")
```

```python
import re

def limpiar_correos(lista_correos):
    # Solo permite letras, números y símbolos estándar de correos
    # Elimina tildes, eñes y caracteres invisibles
    limpios = []
    for email in lista_correos:
        # Forzar a string y quitar espacios
        email = str(email).strip()
        # Filtrar solo caracteres ASCII válidos
        email_limpio = email.encode("ascii", "ignore").decode("ascii")
        if email_limpio == email:
            limpios.append(email)
        else:
            print(f"⚠️ Correo corregido/ignorado por caracteres raros: {email} -> {email_limpio}")
            limpios.append(email_limpio)
    return limpios

# Limpiamos los bloques en memoria
bloques[1] = limpiar_correos(bloques[1]) # Bloque 2
bloques[2] = limpiar_correos(bloques[2]) # Bloque 3
```
