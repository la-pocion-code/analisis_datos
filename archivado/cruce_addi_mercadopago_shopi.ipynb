```python
import pandas as pd
import numpy as np
import zipfile
from pathlib import Path

ruta = rc.validar_ruta()

ruta_addi = ruta / 'conciliaciones_ecommerce' / 'addi'
ruta_salida = ruta / 'conciliaciones_ecommerce' / 'resultados' / 'addi'
ruta_salida.mkdir(parents=True, exist_ok=True)

dfs = []

# 🔎 Buscar archivos zip
archivos_zip = list(ruta_addi.glob("*.zip"))

if archivos_zip:
    print(f"Se encontraron {len(archivos_zip)} archivos ZIP")

    for archivo_zip in archivos_zip:
        print(f"\nProcesando archivo: {archivo_zip.name}")

        if not zipfile.is_zipfile(archivo_zip):
            print("No es un zip válido, se omite.")
            continue

        with zipfile.ZipFile(archivo_zip, 'r') as z:

            archivos_internos = [
                f for f in z.namelist()
                if 'pagos/' in f.lower() and 'resumen general.xlsx' in f.lower()
            ]

            if not archivos_internos:
                print("No se encontró 'resumen general.xlsx'")
                continue

            with z.open(archivos_internos[0]) as archivo_excel:

                df = pd.read_excel(
                    archivo_excel,
                    sheet_name='Transacciones + cancelaciones',
                    engine="calamine"
                )

                # Limpieza
                df = df[~df[df.columns[1]].isna()]
                df.columns = df.iloc[0].values
                df = df.iloc[1:]

                dfs.append(df)



else:
    print("⚠️ No se encontraron archivos ZIP. Buscando CSV...")
    
    archivos_csv = list(ruta_addi.glob("*.csv"))

    if archivos_csv:
        print(f"Se encontraron {len(archivos_csv)} CSV")
        dfs = rc.consolidar_carpeta(ruta_carpeta=ruta_addi, extension='csv', sep=',')
    else:
        raise FileNotFoundError("No se encontraron archivos ZIP ni CSV en la carpeta.")
```
